```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('/content/sample_data/road.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Canny Edge Detection
edges = cv2.Canny(image, 50, 150)

# Display result
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.show()
```

```python
image = cv2.imread('/content/sample_data/road.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply threshold
ret, thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# Display result
plt.imshow(thresh, cmap='gray')
plt.title('Region-based Segmentation')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray, 50, 150)

# Detect lines using Hough Transform
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 68, minLineLength=15, maxLineGap=250)

# Draw lines on the image
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Display result
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Hough Line Transform')
plt.show()
```

```python
# Load image
image = cv2.imread('/content/sample_data/road.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect circles using Hough Transform
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=0, maxRadius=0)

# Draw circles on the image
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(image, (i[0], i[1]), i[2], (0, 255, 0), 2)

# Display result
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Hough Circle Transform')
plt.show()
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
import matplotlib.pyplot as plt

# Load Image
image_path = "/content/sample_data/road.jpg"
image = cv2.imread(image_path)

# Check if image loaded correctly
if image is None:
    raise FileNotFoundError(f"Error: Unable to load image at {image_path}. Check the path and file format.")

# Convert image to RGB for visualization
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Initialize Selective Search
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
ss.setBaseImage(image)
ss.switchToSelectiveSearchFast()  # Faster but fewer proposals

# Get Region Proposals
rects = ss.process()
print(f"Total Region Proposals: {len(rects)}")

# Load Pretrained VGG16 Model
model = VGG16(weights="imagenet", include_top=True)

# Process each ROI
top_n = 20  # Number of ROIs to process (adjust for performance)
detections = []

for i, (x, y, w, h) in enumerate(rects[:top_n]):
    roi = image[y:y+h, x:x+w]  # Extract ROI

    # Resize ROI to VGG16's input size (224x224)
    roi_resized = cv2.resize(roi, (224, 224))
    roi_array = np.expand_dims(roi_resized, axis=0)
    roi_array = preprocess_input(roi_array)  # Normalize

    # Perform Prediction
    predictions = model.predict(roi_array)
    decoded_predictions = decode_predictions(predictions, top=1)[0]  # Get top-1 prediction

    label, confidence = decoded_predictions[0][1], decoded_predictions[0][2]

    # Filter high-confidence detections (threshold: 50%)
    if confidence > 0.50:
        detections.append((x, y, w, h, label, confidence))

# Draw Detected Objects on Image
for (x, y, w, h, label, confidence) in detections:
    cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(image_rgb, f"{label} ({confidence*100:.2f}%)", (x, y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

# Show Image with Detections
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("Object Detection using RCNN (VGG16 + Selective Search)")
plt.show()
```
