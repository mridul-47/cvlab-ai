```python
from google.colab import files
import cv2
import numpy as np

# Upload the image
uploaded = files.upload()

# Read the uploaded image
for filename in uploaded.keys():
    img = cv2.imdecode(np.frombuffer(uploaded[filename], np.uint8), cv2.IMREAD_COLOR)
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt

# Compute histogram for grayscale image
gray_hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

# Plot grayscale histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Pixel Count")
plt.plot(gray_hist, color='black')
plt.xlim([0, 256])
plt.show()
```

```python
# Compute and plot histograms for each color channel
colors = ('b', 'g', 'r')
plt.figure()
plt.title("Color Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Pixel Count")

for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)  # Plot histogram with respective color
    plt.xlim([0, 256])

plt.show()
```

```python
# Apply histogram equalization
equalized_img = cv2.equalizeHist(gray_img)

# Display original and equalized images
plt.figure(figsize=(10,5))

# Original grayscale image
plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

# Equalized image
plt.subplot(1,2,2)
plt.title("Histogram Equalized")
plt.imshow(equalized_img, cmap='gray')
plt.axis("off")

plt.show()
```
